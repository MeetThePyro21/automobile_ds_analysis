```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Обработа данных

```python
df = pd.read_csv('D:/Ivan_Mishin/dataset.csv')
df
```

```python
df['price'] = df['price'].replace('?',np.nan)
df['price'] = pd.to_numeric(df['price'])
df = df[df['price'].notna()]
```

```python
col_change = ['normalized-losses', 'bore',  'stroke', 'horsepower', 'peak-rpm']
for col in col_change:
    df[col] = df[col].replace('?', np.nan)
    df[col] = pd.to_numeric(df[col])
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
df['num-of-doors'] = df['num-of-doors'].replace('?',np.nan)
df = df[df['num-of-doors'].notna()]
```

```python
df['normalized-losses']
```

```python
df['normalized-losses'] = df['normalized-losses'].replace(256.0, np.nan)
df['normalized-losses'] = df['normalized-losses'].replace(231.0, np.nan)
df['normalized-losses'].fillna(df['normalized-losses'].mean(), inplace=True)
```

```python
df['peak-rpm'].unique()
```

```python
df['peak-rpm'] = df['peak-rpm'].replace(6600.0, np.nan)
df['peak-rpm'].fillna(df['peak-rpm'].mean(), inplace=True)
```

### Есть столбцы, которые можно перевести из строк в числа (кол-во дверей, цилиндров)

```python
new_columns = {"num-of-doors":     {"four": 4, "two": 2},
                "num-of-cylinders": {"four": 4, "six": 6, "five": 5, "eight": 8,
                                  "two": 2, "twelve": 12, "three":3 }}
df = df.replace(new_columns)
df.head()
```

### переведем марки в машины в числовой вид (с помощью  OrdinalEncoder)

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
df["make_num"] = ord_enc.fit_transform(df[["make"]])
df[["make", "make_num"]].sample(10)
```

### переведем столбецы "привод", "тип двигателя", "форма машины" в числовой формат

```python
df = pd.get_dummies(df, columns=["body-style", "drive-wheels", "fuel-type"], prefix=["body", "drive", 'fuel'])
df.head()
```

```python
df.select_dtypes(include=np.number).columns.tolist()
```

```python
df_num = df[['symboling', 'normalized-losses','num-of-doors', 'wheel-base', 'length', 'width',
             'height','curb-weight', 'num-of-cylinders', 'engine-size', 'bore', 'stroke',
             'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price',
             'make_num', 'body_convertible', 'body_hardtop', 'body_hatchback', 'body_sedan',
             'body_wagon', 'drive_4wd', 'drive_fwd', 'drive_rwd', 'fuel_diesel', 'fuel_gas']]
df_num.info()
```

# Создание ML модели

```python
df_for_ml= df_num.drop('price',axis = 1)
df_for_ml.info()
```

```python
df_price = df_num['price']
df_price.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df_for_ml, df_price,
                                                    test_size=0.10, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

### Линейная регрессия 

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(x_train, y_train)
predictions = reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score

print("r2_score is : " , r2_score(y_test, predictions))
```

```python
y_test_arr = y_test.to_numpy()
y_test_arr
```

```python
predictions
```

```python
sns.regplot(x = y_test, y = predictions)
```
